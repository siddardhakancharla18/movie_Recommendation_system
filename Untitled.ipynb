```python
import ast
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
            L.append(i['name'])
    return L 
```

```python
def convert(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter+=1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies['crew'] = movies['crew'].apply(convert3)
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(lambda x: x[0:3])
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import nltk
import spacy
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
nlp = spacy.load('en_core_web_sm')
def preprocess(text):
    text = text.lower()
    #remove special characters
    text = re.sub(r'[^a-zA-Z\s]',' ',text)
    text = re.sub(r'\s+',' ',text).strip()

    #tokenization with spacy
    doc = nlp(text)
    token = [word.text for word in doc]
    #remove stopwords
    token = [word for word in token if word not in stop_words]
    #stemming
    stem = [stemmer.stem(word) for word in token]

    return ' '.join(stem)
    
```

```python
movies.head()
```

```python
movies['cleaned'] = movies['all'].apply(preprocess)
```

```python
movies.head()
```

```python
tfid = TfidfVectorizer(max_features = 5000)
vector = tfid.fit_transform(movies['cleaned'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vector,vector)
```

```python
similar[1]
```

```python
def recommend(movie):
    movie = movie.lower()
    index = movies[movies['title'].str.lower() == movie]
    if index.empty:
        print(f"Movie '{movie}' not found in dataset.")
        return

    index = index.index[0]
    distances = sorted(list(enumerate(similar[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("the avengers")
```
